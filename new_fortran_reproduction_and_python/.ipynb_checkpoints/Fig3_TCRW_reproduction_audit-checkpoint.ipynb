3. Plot compact versions of the scalar panels: (a), (b), (e), (f), (g), (j).

```python
summary_files = {
    "a": "tcrw_fig3a_summary.txt",
    "b": "tcrw_fig3b_summary.txt",
    "cde": "tcrw_fig3cde_summary.txt",
    "e": "tcrw_fig3e_summary.txt",
    "f": "tcrw_fig3f_summary.txt",
    "g": "tcrw_fig3g_summary.txt",
    "hij": "tcrw_fig3hij_summary.txt",
    "j": "tcrw_fig3j_summary.txt",
}

data = {key: load_summary(path) for key, path in summary_files.items()}

rows = []
for key, arr in data.items():
    x = arr[:, 1]
    rows.append((key, arr.shape[0], arr.shape[1], float(np.min(x)), float(np.max(x))))

print("key  rows  cols  x_min        x_max")
for key, nrow, ncol, xmin, xmax in rows:
    print(f"{key:3s}  {nrow:4d}  {ncol:4d}  {xmin:10.5g}  {xmax:10.5g}")
```

## Exact-vs-Fortran Vector Audit For Panels (c), (d), (h), and (i)

The exact field comes from the authors' transition matrix through `tcrw_fig3_exact.py`.
The MC field comes from the current Fortran summary files. Angle residuals are computed as wrapped differences in radians.

Panel (h) has a special mask around omega = 0.5 because $|J_{D_r}|$ nearly vanishes there, making direction physically ill-defined.


```python
t0 = time.time()
mc_cde = ov._load_fortran_scan("tcrw_fig3cde_summary.txt")
mc_hij = ov._load_fortran_scan("tcrw_fig3hij_summary.txt")
exact_cde = ov._exact_cde_scan(mc_cde.x, omega=1.0)
exact_hij = ov._exact_hij_scan(mc_hij.x, D_r_fixed=1.0e-3)
print(f"Exact fields computed in {time.time() - t0:.1f} s")
```

```python
def angle_residual(exact, mc, kind, min_exact_fraction=1e-12):
    Jx_e, Jy_e, mag_e = ov._components(exact, kind)
    Jx_m, Jy_m, mag_m = ov._components(mc, kind)
    th_e = np.arctan2(Jy_e, Jx_e)
    th_m = np.arctan2(Jy_m, Jx_m)
    valid = (mag_e > max(float(np.nanmax(mag_e)) * min_exact_fraction, 1e-300)) & (mag_m > 0)
    dtheta = np.full_like(mag_e, np.nan, dtype=float)
    dtheta[valid] = angle_diff(th_m[valid], th_e[valid])
    return dtheta, valid

def summarize_residual(name, exact, mc, kind, min_exact_fraction=1e-12):
    dtheta, valid = angle_residual(exact, mc, kind, min_exact_fraction=min_exact_fraction)
    vals = np.abs(dtheta[valid])
    flat_valid = np.argwhere(valid)
    iy, ix = flat_valid[int(np.nanargmax(vals))]
    return {
        "panel": name,
        "used": int(vals.size),
        "skipped": int(valid.size - vals.size),
        "median_rad": float(np.nanmedian(vals)),
        "p90_rad": float(np.nanpercentile(vals, 90)),
        "max_rad": float(np.nanmax(vals)),
        "worst_x": float(mc.x[ix]),
        "worst_y": int(ov.YS[iy]),
    }

stats = [
    summarize_residual("c: J_Dr vs D_r", exact_cde, mc_cde, "Dr"),
    summarize_residual("d: J_omega vs D_r", exact_cde, mc_cde, "om"),
    summarize_residual("h: J_Dr vs omega", exact_hij, mc_hij, "Dr", min_exact_fraction=5e-3),
    summarize_residual("i: J_omega vs omega", exact_hij, mc_hij, "om"),
]

header = "| panel | used | skipped | median rad | p90 rad | max rad | worst x | worst y |\n|---|---:|---:|---:|---:|---:|---:|---:|"
lines = [header]
for s in stats:
    lines.append(
        f"| {s['panel']} | {s['used']} | {s['skipped']} | "
        f"{s['median_rad']:.3e} | {s['p90_rad']:.3e} | {s['max_rad']:.3e} | "
        f"{s['worst_x']:.5g} | {s['worst_y']} |"
    )
display(Markdown("\n".join(lines)))
```

```python
ix = 0  # smallest D_r
th_exact = np.arctan2(exact_cde.Jy_om[:, ix], exact_cde.Jx_om[:, ix])
th_mc = np.arctan2(mc_cde.Jy_om[:, ix], mc_cde.Jx_om[:, ix])
dtheta = angle_diff(th_mc, th_exact)

print(f"smallest D_r = {mc_cde.x[ix]:.5e}")
print(f"exact theta range = {th_exact.min():.6f} .. {th_exact.max():.6f} rad")
print(f"MC theta range    = {th_mc.min():.6f} .. {th_mc.max():.6f} rad")
print(f"max |MC-exact|    = {np.max(np.abs(dtheta)):.3e} rad")

fig, ax = plt.subplots(figsize=(7.2, 3.4))
ax.axhline(np.pi / 4, color="0.25", lw=1.2, label=r"exact $\pi/4$")
ax.plot(ov.YS, th_mc, "o-", color="C3", label="Fortran MC")
ax.set_xlabel("left-wall y")
ax.set_ylabel(r"$\theta_{J_\omega}$ at smallest $D_r$ (rad)")
ax.set_title("Panel (d): small-D_r roughness is MC current noise")
ax.legend()
fig.tight_layout()
```

```python
residual_specs = [
    ("(c) J_Dr", exact_cde, mc_cde, "Dr", mc_cde.x, "log", 1e-12),
    ("(d) J_omega", exact_cde, mc_cde, "om", mc_cde.x, "log", 1e-12),
    ("(h) J_Dr", exact_hij, mc_hij, "Dr", mc_hij.x, "linear", 5e-3),
    ("(i) J_omega", exact_hij, mc_hij, "om", mc_hij.x, "linear", 1e-12),
]

fig, axs = plt.subplots(2, 2, figsize=(12.4, 6.6))
for ax, (title, exact, mc, kind, xvals, xmode, min_frac) in zip(axs.ravel(), residual_specs):
    dtheta, valid = angle_residual(exact, mc, kind, min_exact_fraction=min_frac)
    image = np.abs(dtheta)
    im = ax.imshow(
        image,
        origin="lower",
        aspect="auto",
        interpolation="nearest",
        cmap="magma",
        vmin=0.0,
        vmax=0.6,
        extent=[0, len(xvals) - 1, ov.YS[0], ov.YS[-1]],
    )
    tick_idx = np.linspace(0, len(xvals) - 1, 5).round().astype(int)
    ax.set_xticks(tick_idx)
    if xmode == "log":
        ax.set_xticklabels([f"{xvals[i]:.0e}" for i in tick_idx])
    else:
        ax.set_xticklabels([f"{xvals[i]:.2g}" for i in tick_idx])
    ax.set_xlabel(r"$D_r$" if xmode == "log" else r"$\omega$")
    ax.set_ylabel("left-wall y")
    ax.set_title(title)
    fig.colorbar(im, ax=ax, fraction=0.046, pad=0.025, label=r"$|\Delta\theta|$ rad")
fig.suptitle("Angle-residual heatmaps: Fortran MC minus exact steady state")
fig.tight_layout()
```

- Panels (c), (h), and (i) agree well between Fortran MC and exact steady state wherever the current direction is meaningful.
